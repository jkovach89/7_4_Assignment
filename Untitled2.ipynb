```python
#! python3
# Module 7.4 Data Visualization Assignment

import numpy as np, pandas as pd, matplotlib.pyplot as plt, os
%matplotlib inline
# Get correct directory
os.chdir('/Users/JKovach/Documents/test projects/duptest13_Thinkful_7_4_Assignment')

# import data set as file object
min_wage = pd.read_csv('./Minimum Wage Data.csv', encoding = "ISO-8859-1")
print(min_wage)
```

```python
# Create new dataframe variables to isolate certain sets
ca_plot = min_wage.loc[lambda ca: min_wage['State']=='California']
fed_plot = min_wage.loc[lambda fed: min_wage['State']=='Federal (FLSA)']

# Create plot of absolute hourly wage vs. 2018 adjusted hourly wage
plt.plot(ca_plot['High.Value'], color='green')
plt.plot(ca_plot['High.2018'], color='blue')
plt.plot(fed_plot['High.Value'], color='red')
plt.plot(fed_plot['High.2018'], color='orange')
plt.show()

# This plot shows the value of minimum wages federally and in california relative to their adjusted 2018 value. 
# As expected, warm color lines and cool color lines converge as they get closer to present day.
# California has stayed congruent with the federal minimum until about the 1700 index = 1998
# I had trouble getting the years to display as the index in the line plot without affecting the data
```

```python
# Create new dataframe variables to isolate certain sets
mean_plot = min_wage.groupby('Year').aggregate(np.mean)

plt.scatter(x=ca_plot['Year'], y=ca_plot['High.Value'], color='blue')
plt.scatter(x=fed_plot['Year'], y=fed_plot['High.Value'], color='red')
plt.scatter(x=mean_plot.index, y=mean_plot['High.Value'], color='purple')
plt.show()

# Scatter plotting was much more amenable to specifying the x axis. 
# The scatter plot below shows the national average(purple) compared to federal(red) and California(blue)
# Surprisingly, the national average almost always sits below the federal. This could be due to several factors:
#      1. The national average includes the federal data which could affect the average figure
#      2. Some states, such as alabama, do not have a minimum wage and data for such was reflected as '0'
# Interestingly, California had a minimum wage below the federal from 1976-1978, and again in 1980
```

```python
state_plot = min_wage.groupby('State').aggregate(np.mean)
plot_1968 = min_wage.loc[lambda se: min_wage['Year']==1968]
plot_2017 = min_wage.loc[lambda se: min_wage['Year']==2017]

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.boxplot(np.array(list(state_plot['High.Value'])))
plt.ylabel('Distribution: 50 year average')

plt.subplot(2, 2, 2)
plt.boxplot(np.array(list(plot_1968['High.Value'])))
plt.ylabel('Distribution: 1968')

plt.subplot(2, 2, 3)
plt.boxplot(np.array(list(plot_2017['High.Value'])))
plt.ylabel('Distribution: 2017')

plt.show()

# The box plots below show state average wage from 1968 to 2017, 1968, and 2017, respectively.
# For some reason 1968 will not chart. See error message.
# As expected, 2017 wages have a higher central tendency than the 50 year average, 1968 is expected to be lower.
# Interestingly, less outliers are present in the 2017 distribution, indicating states have increases wages,
# or been forced to increase due to rising federal wages
```

```python
plt.hist(min_wage['High.2018'], bins=50, color='blue', alpha=.5)
plt.hist(min_wage['High.Value'], bins=50, color='red', alpha=.5)
plt.show()

# The histogram below shows the distribution of historical wages(red) versus their adjusted 2018 values(blue).
# As expected, adjusted values seem to center around the 7.25 figure (current federal minimum),
# while historical figures are more evenly distributed between $1-$5.
# Confirming the inference in chart 2, we see that the majority of values fall in the '0' bin, meaning
# either data was unavailable, or the state did not have a state-mandated minimum wage.
```
